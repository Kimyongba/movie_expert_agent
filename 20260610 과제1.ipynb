```python
import openai
import requests

client = openai.OpenAI()

BASE_URL = "https://nomad-movies-2.nomadcoders.workers.dev"
```

```python
def get_popular_movies():
    return requests.get(f"{BASE_URL}/movies").json()

def get_movie_details(id):
    return requests.get(f"{BASE_URL}/movies/{id}").json()

def get_movie_credits(id):
    return requests.get(f"{BASE_URL}/movies/{id}/credits").json()
```

```python
def ask_movie_agent(question):
    PROMPT = f"""
I have the following functions in my system.

`get_popular_movies`
`get_movie_details`
`get_movie_credits`

`get_popular_movies` receives no arguments.
`get_movie_details` receives a movie id as an argument. i.e get_movie_details(550)
`get_movie_credits` receives a movie id as an argument. i.e get_movie_credits(550)

Please answer with the name of the function that you would like me to run.

Please say nothing else, just the name of the function with the arguments.

Answer the following question:

{question}
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": PROMPT}],
    )

    message = response.choices[0].message.content
    return message
```

```python
message = ask_movie_agent("지금 인기 있는 영화가 무엇인지 알려줘")
message
```

```python
import openai

client = openai.OpenAI()
messages = []

SYSTEM_PROMPT = """
You are a friendly movie recommendation chatbot.

You must remember the user's movie preferences from the conversation history.
You must remember the movies the user already watched.
When recommending movies, use the user's favorite genre and do not recommend movies the user already watched.
When the user asks what they said before, answer based on the conversation history.

Answer in Korean.
"""
```

```python
def call_ai():
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            *messages,
        ],
    )
    message = response.choices[0].message.content
    messages.append({"role": "assistant", "content": message})
    print(f"AI: {message}")
```

```python
def send_message(message):
    messages.append({"role": "user", "content": message})
    print(f"User: {message}")
    call_ai()
```

```python
send_message("나는 SF 영화를 좋아해")
```

```python
send_message("인셉션이랑 인터스텔라는 이미 봤어")
```
